# 1. Import Libraries and Load the Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/titanic/Titanic-Dataset.csv")
```

# 2. Data Overview

```python
# View basic info and summary statistics
df.info()
df.describe(include="all")
# Check for missing values
df.isnull().sum()
```

# 3. Visualize Numeric Features

```python
# Histograms
df['Age'].plot.hist(bins=30, edgecolor='black', alpha=0.7)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()

df['Fare'].plot.hist(bins=30, edgecolor='black', alpha=0.7)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.show()

# Boxplots
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Survival by Age")
plt.show()

sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare by Passenger Class")
plt.show()
```

# 4. Visualize Categorical Features

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival Count by Sex")
plt.show()

sns.countplot(x="Embarked", hue="Survived", data=df)
plt.title("Survival Count by Embarkation Port")
plt.show()

sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival Count by Passenger Class")
plt.show()
```

# 5. Feature Relationships

```python
# Correlation matrix
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Feature Correlation Matrix")
plt.show()

# Pairplot
sns.pairplot(df, hue="Survived", vars=["Age","Fare","Pclass","SibSp","Parch"])
plt.suptitle("Pairplot of Numerical Features", y=1.01)
plt.show()
```
